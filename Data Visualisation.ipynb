```python
#types
#univariate, bivariate and multivariate

#univariate means only one variable at a time, histograms barcharts piecharts boxplots can be plotted, categorical or numerical

#bivariate means two at a time, can be numerical or categorical
#scenarios : numerical vs numerical, nmerical vs categorical, categorical vs categorical
#num vs num we use scatterplot
#num vs cat we use ANOVA
#cat vs cat we use Chi Squared error test

#multi variate analysis to analyse many variables, uses clustering, principal component analysis(PCA) and Multi correspondence analysis
#Clustering groups datapoints 
#PCA transforms large dataset to smaller with no to minimal loss of information
#MCA separates based on categories
```

# Pie Chart

```python
labels=["php","java","python","c"]
num=[75,100,80,20]
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.pie(num,labels=labels,autopct="%0.2f%%")
axes1.legend()
```

```python
#=["php","pyhton", "java", "c"]
Explode=[0.5,0,0,0.2]
plt.pie(num, labels=labels, explode=Explode)
```
